# Array Operations

```python
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)
```

- Scalar Operations

```python
# Arithmetic 
a1 * 2 # Mathematical Operators can be used + - / % **
```

```python
# Relational Operators
a2 > 15 # == >= <= > < != !
```

- Vector Operations

```python
a1 + a2 # + - / % * **
```

# Array Functions

```python
a1 = np.random.random((3,3))
a1 = np.round(a1 * 100)
a1
```

- Max / Min / Sum / Product

```python
print(np.max(a1))
print(np.min(a1))
print(np.sum(a1))
print(np.prod(a1))

print(np.min(a1, axis=1)) # For  row/column (axis = 0 -> Column | axis = 1 -> Row)
print(np.sum(a1, axis=1))
```

- Statistcal Operations

```python
# mean, median, std -> Standard Deviation, var -> Variance
np.mean(a1)
np.mean(a1, axis = 1)
```

- Trignometric funtions

```python
np.sin(a1) # All other Trignometric functions
```

- Dot Product

```python
# Simply multiplication of 2 matrices (If the number of cols of one matrix == number of rows of other)
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(12,24).reshape(4,3)

np.dot(a2, a3)
```

- Log and Exponents

```python
np.log(a1)
```

```python
np.exp(a1)
```

# Stacking

```python
# Horizontal Stacking            # Shape must be same for stacking
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)

np.hstack((a4,a5))
```

```python
# Vertical Stacking
np.vstack((a4,a5))
```

# Splitting

```python
# Horizontal Splitting
np.hsplit(a4,2)
```

```python
np.vsplit(a5,3)
```
