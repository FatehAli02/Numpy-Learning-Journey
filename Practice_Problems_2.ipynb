### 9. Normalize a 5x5 random matrix

```python
import numpy as np

matrix = np.random.randint(1,100,25).reshape(5,5)
print(matrix)
```

```python
min = matrix.min()
max = matrix.max()

normal_matrix = (matrix - min) / (max - min)

print(matrix)
print(normal_matrix)

```

### 11. Given a 1D array, negate all elements which are between 3 and 8, in place.

```python
arr = np.arange(1,10)
arr
```

```python
arr[(arr >= 3) & (arr <=8)] *= -1
arr
```

```python
b1 = np.array([2,5,4,8,7])
b2 = np.array([4,8,0,1,3])

np.intersect1d(b1,b2)
```
